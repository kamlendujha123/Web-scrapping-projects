````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
boxes=soup.find('div',class_='title-list-grid')
````

## **Fetching Movie URL's**

````python
# for url this is
urls=[]
movie_url=boxes.find_all('a',class_='title-list-grid__item--link')
for each_url in movie_url[:50]:
  each_url=each_url['href']
  link='https://www.justwatch.com'+each_url
  urls.append(link)
print(urls)
print(len(urls))

````

## **Scrapping Movie Title**

````python
# Write Your Code here
title=[]
names=boxes.find_all('img',class_='picture-comp__img')
for name in names[:50]:
  name=name['alt']
  title.append(name)
print(title)
print(len(title))
````

## **Scrapping release Year**

````python
# Write Your Code here
release_year=[]
for i in urls:
  page1=requests.get(i)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup1=BeautifulSoup(page1.text,'lxml')
  years=soup1.find('span',class_='text-muted')
  release_year.append(years.text.split('(')[-1].split(')')[0])
print(release_year)
print(len(release_year))
````

## **Scrapping Genres**

````python
# Write Your Code here
# Write Your Code here
genres=[]
for i in urls:
  page1=requests.get(i)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup1=BeautifulSoup(page1.text,'html.parser')
  movie_genres=soup1.find('h3',class_='detail-infos__subheading',string='Genres')
  movie_genres1=movie_genres.find_next_sibling('div',class_='detail-infos__value')
  genres.append(movie_genres1.text)
print(genres)
#print(len(genres))
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
imdb_rating=[]

for i in urls:
  try:
# Sending an HTTP GET request to the URL
    page=requests.get(i)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
    soup=BeautifulSoup(page.text,'lxml')
# Printing the prettified HTML content
    ratings=soup.find_all('div',class_="jw-scoring-listing__rating")
    imdb_rating.append(ratings[1].text.split()[0])
  except:
    imdb_rating.append('N/A')
print(imdb_rating)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here for runtime
runtime=[]
for i in urls:
  page1=requests.get(i)
  soup=BeautifulSoup(page1.text,'lxml')
  movie_runtime=soup.find('h3',class_='detail-infos__subheading',string='Runtime')
  runtime1=movie_runtime.find_next_sibling('div',class_='detail-infos__value')
  runtime.append(runtime1.text)
print(runtime)
````

## **Scrapping Age Rating**

````python
# Write Your Code here
age_rating=[]
for i in urls:
  page=requests.get(i)
  soup=BeautifulSoup(page.text,'lxml')
  movie_age_rating=soup.find('h3','detail-infos__subheading',string='Age rating')
  if movie_age_rating:
    age_rating1=movie_age_rating.find_next_sibling('div',class_='detail-infos__value')
    age_rating.append(age_rating1.text)
  else:
    age_rating.append('N/A')
print(age_rating)
print(len(age_rating))
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
production_countries=[]
for i in urls:
  page1=requests.get(i)
  soup=BeautifulSoup(page1.text,'lxml')
  region=soup.find('h3',class_='detail-infos__subheading',string=' Production country ')
  region1=region.find_next_sibling('div',class_='detail-infos__value')
  production_countries.append(region1.text)
print(production_countries)
print(len(production_countries))
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
streaming_service=[]
for i in urls:
  page1=requests.get(i)
  soup=BeautifulSoup(page1.text,'lxml')
  stream=soup.find('img',class_='offer__icon')
  if stream:
    stream1=stream['alt']
    streaming_service.append(stream1)
  else:
    streaming_service.append('Not Avaliable')
print(streaming_service)
````

## **Now Creating Movies DataFrame**

````python

# Write Your Code here
Movies=pd.DataFrame({'urls':urls,'title':title,'release_year':release_year,'imdb_rating':imdb_rating,'genres':genres,'age_rating':age_rating,'streaming_service':streaming_service})
Movies
````

## **Scraping TV  Show Data**

````python
# Specifying the URL from which tv show related data will be fetched
tv_url='https://www.justwatch.com/in/tv-shows?release_year_from=2000'
# Sending an HTTP GET request to the URL
page=requests.get(tv_url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
#print(soup.prettify())
show_boxes=soup.find('div',class_='title-list-grid')
#print(boxes)
````

## **Fetching Tv shows Url details**

````python
# Write Your Code here
urls=[]
url1=show_boxes.find_all('a',class_='title-list-grid__item--link')
for url2 in url1[:50]:
  url2=url2['href']
  new_url='https://www.justwatch.com'+url2
  urls.append(new_url)

print(urls)
````

## **Fetching Tv Show Title details**

````python
# Write Your Code here
title=[]
for i in urls:
  page=requests.get(i)
  soup1=BeautifulSoup(page.text,'lxml')
  title1=soup1.find('div',class_='title-block')
  title2=title1.find('h1')
  title.append(title2.text)
print(title)
print(len(title))
````

## **Fetching Release Year**

## **Fetching TV Show Genre Details**

## **Task 2 :- Data Filtering & Analysis**

# For movies dataframe

````python
# Write Your Code here
# now i don't want to include those movies who's imdb_rating is N/A so i'll use dropna() function
from datetime import datetime,timedelta
Movies['imdb_rating']=pd.to_numeric(Movies['imdb_rating'],errors='coerce')
Movies=Movies.dropna(subset=['imdb_rating'])
filtered_movies=Movies[Movies['imdb_rating']>7]
filtered_movies['release_year']=pd.to_datetime(filtered_movies['release_year'],errors='coerce')
day=datetime.now()-timedelta(days=365*2)
df=filtered_movies[filtered_movies['release_year']>day]
print(df)
print(len(df))
````

# For Web Shows dataframe

````python
# Write Your Code here
from datetime import datetime, timedelta
shows['imdb_rating']=pd.to_numeric(shows['imdb_rating'],errors='coerce')
shows=shows.dropna(subset=['imdb_rating'])
filtered_shows=shows[shows['imdb_rating']>7]
filtered_shows['release_year']=pd.to_datetime(filtered_shows['release_year'],errors='coerce')
days=datetime.now()-timedelta(365*2)
show_list=filtered_shows[filtered_shows['release_year']>days]
print(show_list)
print(len(show_list))
````

## **Calculating Mean IMDB Ratings for both Movies and Tv Shows**

````python
# Write Your Code here
# for average imdb rating for movies
mean_imdb_rating_movie=df.aggregate({'imdb_rating':'mean'})
print(mean_imdb_rating_movie)
print()
# for average imdb rating for shows
mean_imdb_rating_show=show_list.aggregate({'imdb_rating':'mean'})
print(mean_imdb_rating_show)
````
